```python
#TABLE BRING

# Bring the BETTING table fom SQL
query_betting = "SELECT * FROM betting"
df_betting = pd.read_sql(query_betting,cnx)

# 1. Data Loading

df_betting.info()
df_betting.columns
```

```python
df_betting.head (14)
```

```python
print(f"‚ÑπÔ∏è Data info: {df_betting.shape[0]} rows, {df_betting.shape[1]} columns")
print(f"üìä Initial columns: {list(df_betting.columns)}")
```

```python
 # 2. IDENTIFY THE COLUMNS THAT SHOULD BE CONSERVED
columns_to_keep = [
    'AccountNo', 'BetDate', 'ClassId', 'CategoryId', 'Source', 
    'BetCount', 'Bet_Amt', 'Win_Amt', 'Product'
]

 # Aplicated the filter to df to keep only the 9 columns
df_betting = df_betting [ columns_to_keep]

# Remove columns that completly empty
empty_cols = df_betting.columns[df_betting.isnull().all()]

df_betting = df_betting.drop(columns = empty_cols)

print(f"‚úÖ Clean columns. We have:{len(df_betting.columns)} columns now ")

 # Remove Duplicate/ Cleanup: excluded placeholder or error columns like MyUnknownColumn or #N/A.

df_betting_clean = df_betting.drop_duplicates() # Delete duplicate rows
print(f"Duplicate rows removed")

 # Null Value Processing (dropna)

# - If a bet does not have an [AccountNo] ,We cannot charge them or analyze their behavior.
# - If a bet does not have a [BetDate], we do not know when it occurred. We cannot see trends by month or year.
   
# We remove rows where the bet date is null [BetDate]

df_betting_clean = df_betting.dropna(subset=["BetDate"])

print(f"Null in BetDate: {df_betting_clean['BetDate'].isna().sum()}")

# Same thinf with [AccountNo]
df_betting_clean = df_betting.dropna(subset=["AccountNo"])
print(f"Null in : {df_betting_clean['AccountNo'].isna().sum()}")


```

```python
df_betting_clean.head(17)
```

```python
#convert date column to date time
df_betting_clean['BetDate'] = pd.to_datetime(df_betting_clean ['BetDate'],
                                             format = '%d/%m/%Y', errors ='coerce')
```

```python
df_betting_clean.dtypes
df_betting_clean.info()
```

```python

df_betting = df_betting_clean
df_betting.info()
df_betting.head()
```

```python

df_betting = df_betting_clean

# Get the range of date (starting, finishing)

s_date = df_betting['BetDate'].min()
f_date = df_betting['BetDate'].max()
print (s_date)
print(f_date )

#Analysis of betting behavior activity by product.


result_analysis = df_betting.groupby(['BetDate', 'Product'])['Bet_Amt'].sum().unstack()

result_analysis = result_analysis .fillna(0)

result_analysis  = result_analysis .drop(columns=['#N/A'])


# Rmove rows the sum of all columns is 0.
result_analysis = result_analysis[result_analysis.sum(axis=1) > 0]

# Grouping by month (M), add up and transpose to have the months as columns.

result_analysis = result_analysis.resample('M').sum().T

# Rename the columns to name of month.
result_analysis.columns = ['September', 'October', 'November', 'December']


result_analysis
```

```python
result_analysis.info()
```

```python
result_analysis = result_analysis.sort_values(by=['September', 'October', 'November', 'December'], ascending=False)

result_analysis. round(2)

```

```python

import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

result_analysis.T.plot(kind='bar', figsize=(12, 6))

# Add titles and labels
plt.title('Betting Behavior by Product (Sept - Dec 2012)', fontsize=15)
plt.xlabel('Month', fontsize=12)
plt.ylabel(' Total (Float)', fontsize=12)
plt.xticks(rotation=0) 
plt.legend(title='Productos')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
sql_ranking = """
SELECT 
    p.product, 
    SUM(IFNULL(b.Bet_Amt, 0)) AS total_bet -- Cambiamos nulos por 0
FROM product p
LEFT JOIN betting b 
    ON b.ClassId = p.CLASSID 
    AND b.CategoryId = p.CATEGORYID
GROUP BY p.product
ORDER BY total_bet DESC;"""

df_ranking = pd.read_sql(sql_ranking, cnx)
df_ranking['total_bet'] = df_ranking['total_bet'].round(2)
df_ranking.head(8)
print (df_ranking.head(8))
```

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(10, 6))
plt.bar(df_ranking['product'], df_ranking['total_bet'], color='skyblue')
plt.title('Ranking Total de Ganancias por Producto')
plt.xticks(rotation=45)
plt.show()
```

```python
sql_rentabilidad = """
SELECT 
    p.product, 
    SUM(b.Bet_Amt) AS total_bet,
    SUM(b.Win_Amt) AS total_win
FROM betting b
JOIN product p 
    ON b.ClassId = p.CLASSID 
    AND b.CategoryId = p.CATEGORYID
GROUP BY p.product;
"""
```

```python
df_renta = pd.read_sql(sql_rentabilidad, cnx)

df_renta['profitability'] = df_renta['total_bet'] - df_renta['total_win']

df_renta = df_renta.sort_values(ascending=False, by='profitability')

df_renta.head(8)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df_renta['product'], df_renta['profitability'], color='forestgreen')

plt.title('Ranking de Rentabilidad Real por Producto (Profit)', fontsize=14)
plt.ylabel('Ganancia Neta para la Empresa (‚Ç¨)')
plt.xticks(rotation=45)
plt.show()

# Mostramos la tabla para que veas los n√∫meros
print(df_renta[['product', 'total_bet', 'profitability']])
```

```python
# 1. Configuramos el tama√±o de la gr√°fica
plt.figure(figsize=(10, 6))

# 2. Creamos la gr√°fica de barras de Rentabilidad
# Usamos el color 'forestgreen' para representar las ganancias netas
plt.bar(df_renta['product'], df_renta['profitability'], color='forestgreen')

# 3. A√±adimos t√≠tulos y etiquetas claras
plt.title('Ranking de Rentabilidad Real por Producto (Profit)', fontsize=14, fontweight='bold')
plt.xlabel('Productos', fontsize=12)
plt.ylabel('Ganancia Neta para la Empresa (‚Ç¨)', fontsize=12)

# 4. Rotamos los nombres para que se lean perfectamente
plt.xticks(rotation=45, ha='right')

# 5. A√±adimos una cuadr√≠cula suave para facilitar la lectura de los montos
plt.grid(axis='y', linestyle='--', alpha=0.6)

# 6. Ajustamos el dise√±o para que no se corte nada
plt.tight_layout()

# 7. Mostramos la gr√°fica
plt.show()

# --- Extra: Mostramos la tabla resumen debajo ---
print("-" * 30)
print("RESUMEN DE RENTABILIDAD")
print("-" * 30)
print(df_renta[['product', 'total_bet', 'profitability']])
```

```python
import matplotlib.pyplot as plt

# 1. Datos
labels = df_renta['product']
sizes = df_renta['profitability']

# 2. Explosi√≥n para dar aire entre piezas
explode = [0.12] * len(labels) 

# 3. Configuraci√≥n
plt.figure(figsize=(12, 10))
colors = ['#2ca02c', '#1f77b4', '#ff7f0e', '#d62728', '#9467bd', '#8c564b', '#e377c2']

# 4. Crear el disco
patches, texts, autotexts = plt.pie(
    sizes, 
    labels=labels, 
    explode=explode, 
    autopct='%1.1f%%', 
    startangle=160, 
    colors=colors,
    labeldistance=1.25,  # Aleja los nombres (Bingo, Adjustments)
    pctdistance=1.1,     # <--- TRUCO: Aleja los PORCENTAJES fuera del disco
)

# 5. Estilo para que los porcentajes se vean mejor afuera
for autotext in autotexts:
    autotext.set_fontsize(10)
    autotext.set_weight('bold')
    autotext.set_color('black') # Los ponemos negros porque ahora est√°n sobre el fondo blanco

# 6. T√≠tulo
plt.title('Rentabilidad Real: An√°lisis Detallado', fontsize=16, fontweight='bold', pad=40)

# 7. C√≠rculo central (Donut)
centre_circle = plt.Circle((0,0), 0.50, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.axis('equal') 
plt.show()
```
