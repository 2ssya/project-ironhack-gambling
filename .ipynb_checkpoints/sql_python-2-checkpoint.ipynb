```python
#TABLE BRING

# Bring the BETTING table fom SQL
query_betting = "SELECT * FROM betting"
df_betting = pd.read_sql(query_betting,cnx)

# 1. Data Loading

df_betting.info()
df_betting.columns
```

```python
df_betting.head (14)
```

```python
print(f"‚ÑπÔ∏è Data info: {df_betting.shape[0]} rows, {df_betting.shape[1]} columns")
print(f"üìä Initial columns: {list(df_betting.columns)}")
```

```python
 # 2. IDENTIFY THE COLUMNS THAT SHOULD BE CONSERVED
columns_to_keep = [
    'AccountNo', 'BetDate', 'ClassId', 'CategoryId', 'Source', 
    'BetCount', 'Bet_Amt', 'Win_Amt', 'Product'
]

 # Aplicated the filter to df to keep only the 9 columns
df_betting = df_betting [ columns_to_keep]

# Remove columns that completly empty
empty_cols = df_betting.columns[df_betting.isnull().all()]

df_betting = df_betting.drop(columns = empty_cols)

print(f"‚úÖ Clean columns. We have:{len(df_betting.columns)} columns now ")

 # Remove Duplicate/ Cleanup: excluded placeholder or error columns like MyUnknownColumn or #N/A.

df_betting_clean = df_betting.drop_duplicates() # Delete duplicate rows
print(f"Duplicate rows removed")

 # Null Value Processing (dropna)

# - If a bet does not have an [AccountNo] ,We cannot charge them or analyze their behavior.
# - If a bet does not have a [BetDate], we do not know when it occurred. We cannot see trends by month or year.
   
# We remove rows where the bet date is null [BetDate]

df_betting_clean = df_betting.dropna(subset=["BetDate"])

print(f"Null in BetDate: {df_betting_clean['BetDate'].isna().sum()}")

# Same thinf with [AccountNo]
df_betting_clean = df_betting.dropna(subset=["AccountNo"])
print(f"Null in : {df_betting_clean['AccountNo'].isna().sum()}")


```

```python
df_betting_clean.head(17)
```

```python
#convert date column to date time
df_betting_clean['BetDate'] = pd.to_datetime(df_betting_clean ['BetDate'],
                                             format = '%d/%m/%Y', errors ='coerce')
```

```python
df_betting_clean.dtypes
df_betting_clean.info()
```

```python

df_betting = df_betting_clean
df_betting.info()
df_betting.head()
```

```python

df_betting = df_betting_clean

# Get the range of date (starting, finishing)

s_date = df_betting['BetDate'].min()
f_date = df_betting['BetDate'].max()
print (s_date)
print(f_date )

#Analysis of betting behavior activity by product.


result_analysis = df_betting.groupby(['BetDate', 'Product'])['Bet_Amt'].sum().unstack()

result_analysis = result_analysis .fillna(0)

result_analysis  = result_analysis .drop(columns=['#N/A'])

#Deleted columns that have many zeros or are not of interest.

result_analysis  = result_analysis .drop(columns=['Adjustments'])

# Rmove rows the sum of all columns is 0.
result_analysis = result_analysis[result_analysis.sum(axis=1) > 0]

# Grouping by month (M), add up and transpose to have the months as columns.

result_analysis = result_analysis.resample('M').sum().T

# Rename the columns to name of month.
result_analysis.columns = ['September', 'October', 'November', 'December']


result_analysis
```

```python
result_analysis.info()
```

```python
result_analysis = result_analysis.sort_values(by=['September', 'October', 'November', 'December'], ascending=False)

result_analysis. round(2)

```

```python

import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

result_analysis.T.plot(kind='bar', figsize=(12, 6))

# Add titles and labels
plt.title('Betting Behavior by Product (Sept - Dec 2012)', fontsize=15)
plt.xlabel('Meses', fontsize=12)
plt.ylabel(' Total (Float)', fontsize=12)
plt.xticks(rotation=0) 
plt.legend(title='Productos')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
