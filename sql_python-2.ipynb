```python
 # 2. IDENTIFY THE COLUMNS THAT SHOULD BE CONSERVED
columns_to_keep = [
    'AccountNo', 'BetDate', 'ClassId', 'CategoryId', 'Source', 
    'BetCount', 'Bet_Amt', 'Win_Amt', 'Product'
]

 # Aplicated the filter to df to keep only the 9 columns
df_betting = df_betting [ columns_to_keep]

# Remove columns that completly empty
empty_cols = df_betting.columns[df_betting.isnull().all()]

df_betting = df_betting.drop(columns = empty_cols)

print(f"‚úÖ Clean columns. We have:{len(df_betting.columns)} columns now ")

 # Remove Duplicate/ Cleanup: excluded placeholder or error columns like MyUnknownColumn or #N/A.

df_betting_clean = df_betting.drop_duplicates() # Delete duplicate rows
print(f"Duplicate rows removed")

 # Null Value Processing (dropna)

# - If a bet does not have an [AccountNo] ,We cannot charge them or analyze their behavior.
# - If a bet does not have a [BetDate], we do not know when it occurred. We cannot see trends by month or year.
   
# We remove rows where the bet date is null [BetDate]

df_betting_clean = df_betting.dropna(subset=["BetDate"])

print(f"Null in BetDate: {df_betting_clean['BetDate'].isna().sum()}")

# Same thinf with [AccountNo]
df_betting_clean = df_betting.dropna(subset=["AccountNo"])
print(f"Null in : {df_betting_clean['AccountNo'].isna().sum()}")


```
